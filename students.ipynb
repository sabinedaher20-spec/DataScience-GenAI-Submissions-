To answer the first question "Does test preparation actually raise scores?" I will ask Gemini to "create a boxplot comparing test preparation course completion against math, reading, and writing scores." This will show whether students who took the course achieved higher results.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Reload the DataFrame to ensure it's available
df = pd.read_csv('/content/students.csv')

score_cols = ['math score', 'reading score', 'writing score']

plt.figure(figsize=(18, 6))
for i, col in enumerate(score_cols):
    plt.subplot(1, 3, i + 1)
    sns.boxplot(x='test preparation course', y=col, data=df, palette='viridis')
    plt.title(f'{col} by Test Preparation Course')
    plt.xlabel('Test Preparation Course')
    plt.ylabel('Score')
plt.tight_layout()
plt.show()
```

The boxplots show that completing the test preparation course is associated with higher and more consistent scores in math, reading, and writing. Students who took the course clearly outperformed those who did not.

To answer the second question "Are there performance gaps based on race/ethnicity or parental education?" I will ask Gemini to "create boxplots comparing race/ethnicity and parental level of education against math, reading, and writing scores."

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Reload the DataFrame to ensure it's available
df = pd.read_csv('/content/students.csv')

score_cols = ['math score', 'reading score', 'writing score']
categorical_cols = ['race/ethnicity', 'parental level of education']

# Boxplots for Race/Ethnicity vs. Scores
plt.figure(figsize=(20, 6))
for i, col in enumerate(score_cols):
    plt.subplot(1, 3, i + 1)
    sns.boxplot(x='race/ethnicity', y=col, data=df, palette='viridis')
    plt.title(f'{col} by Race/Ethnicity')
    plt.xlabel('Race/Ethnicity')
    plt.ylabel('Score')
    plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Boxplots for Parental Level of Education vs. Scores
plt.figure(figsize=(20, 6))
for i, col in enumerate(score_cols):
    plt.subplot(1, 3, i + 1)
    sns.boxplot(x='parental level of education', y=col, data=df, palette='viridis')
    plt.title(f'{col} by Parental Level of Education')
    plt.xlabel('Parental Level of Education')
    plt.ylabel('Score')
    plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The boxplots show clear performance gaps based on parental education. Students with parents who hold college or advanced degrees achieve higher and more consistent scores, while those whose parents have lower education levels tend to score lower. This suggests that parental education positively influences student achievement across all subjects.

To answer the third question, “Does lunch type influence results?”, I asked Gemini to create a bar chart showing the average math, reading, and writing scores for each lunch type.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Reload the DataFrame to ensure it's available
df = pd.read_csv('/content/students.csv')

score_cols = ['math score', 'reading score', 'writing score']

# Calculate the average scores for each lunch type
avg_scores_by_lunch = df.groupby('lunch')[score_cols].mean().reset_index()

# Melt the DataFrame for easier plotting with seaborn
avg_scores_melted = avg_scores_by_lunch.melt(id_vars=['lunch'], var_name='score_type', value_name='average_score')

# Create the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='lunch', y='average_score', hue='score_type', data=avg_scores_melted, palette='viridis')
plt.title('Average Scores by Lunch Type')
plt.xlabel('Lunch Type')
plt.ylabel('Average Score')
plt.ylim(0, 100) # Set y-axis limit for scores
plt.legend(title='Score Type')
plt.tight_layout()
plt.show()
```

The bar chart shows that lunch type clearly influences results. Students with standard lunches score higher on average in math, reading, and writing, suggesting that socioeconomic and nutritional factors play a key role in academic performance.
